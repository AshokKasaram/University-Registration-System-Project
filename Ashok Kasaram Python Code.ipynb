```python
from tkinter import *
from tkinter import ttk, messagebox
from PIL import Image, ImageTk
import mysql.connector

# Connecting to the database
db = mysql.connector.connect(host="localhost", user="root", passwd="123456789", database="UniversityRegistrationSystem")
mycur = db.cursor()

def set_background(window):
    # Load and display the background image
    img = Image.open(R"C:\Users\kasaram ashok\Downloads\UNIVERSITY LOGO.png")  # Replace "UNIVERSITY LOGO.png" with your image file
    img = img.resize((window.winfo_width(), window.winfo_height()), Image.ANTIALIAS)
    background_img = ImageTk.PhotoImage(img)
    background_label = Label(window, image=background_img)
    background_label.image = background_img
    background_label.place(x=0, y=0, relwidth=1, relheight=1)


def error_destroy():
    err.destroy()

def succ_destroy():
    succ.destroy()
    root1.destroy()

def error():
    global err
    err = Toplevel(root1)
    err.title("Error")
    err.geometry("200x100")
    Label(err, text="All fields are required..", fg="red", font="bold").pack()
    Label(err, text="").pack()
    Button(err, text="Ok", bg="white", fg="black", width=8, height=1, command=error_destroy).pack()

def success():
    global succ
    succ = Toplevel(root1)
    succ.title("Success")
    succ.geometry("200x100")
    Label(succ, text="Registration successful...", fg="green", font="bold").pack()
    Label(succ, text="").pack()
    Button(succ, text="Ok", bg="white", fg="black", width=8, height=1, command=lambda: [succ_destroy(), root1.destroy()]).pack()

def register_user():
    username_info = username.get()
    password_info = password.get()
    if username_info == "":
        error()
    elif password_info == "":
        error()
    else:
        sql = "INSERT INTO login (username, password) VALUES (%s, %s)"
        t = (username_info, password_info)
        mycur.execute(sql, t)
        db.commit()
        success()

def registration():
    global root1
    root1 = Toplevel(root)
    root1.title("Registration Portal")
    root1.geometry("300x250")
    global username
    global password
    Label(root1, text="Register your account", bg="grey", fg="black", font="bold", width=300).pack()
    username = StringVar()
    password = StringVar()
    Label(root1, text="").pack()
    Label(root1, text="Username :",).pack()
    Entry(root1, textvariable=username).pack()
    Label(root1, text="").pack()
    Label(root1, text="Password :").pack()
    Entry(root1, textvariable=password, show="*").pack()
    Label(root1, text="").pack()
    Button(root1, text="Register", bg="white", fg="black", command=register_user).pack()

def login():
    global root2
    root2 = Toplevel(root)
    root2.title("Log-In Portal")
    root2.geometry("300x300")
    global username_varify
    global password_varify
    Label(root2, text="Log-In Portal", bg="grey", fg="black", font="bold", width=300).pack()
    username_varify = StringVar()
    password_varify = StringVar()
    Label(root2, text="").pack()
    Label(root2, text="Username :",).pack()
    Entry(root2, textvariable=username_varify).pack()
    Label(root2, text="").pack()
    Label(root2, text="Password :").pack()
    Entry(root2, textvariable=password_varify, show="*").pack()
    Label(root2, text="").pack()
    Button(root2, text="Log-In", bg="white", fg="black", command=login_varify).pack()
    Label(root2, text="")

def logg_destroy():
    logg.destroy()
    root2.destroy()

def fail_destroy():
    fail.destroy()

def logged():
    global logg
    logg = Toplevel(root2)
    logg.title("Welcome")
    logg.geometry("300x200")
    
    # Background color
    logg.configure(bg="#f0f0f0")
    
    # Text color
    text_color = "#333333"
    
    # Welcome message
    Label(logg, text="Welcome {} ".format(username_varify.get()), fg="#008000", font=("Arial", 14, "bold"), bg="#f0f0f0").pack()
    Label(logg, text="", bg="#f0f0f0").pack()
    
    # Log-out button
    Button(logg, text="Log-Out", bg="#cccccc", fg=text_color, width=10, height=2, command=logg_destroy).pack()

    # Buttons for student details
    Button(logg, text="Add Student Details", bg="white", fg="black", command=add_student_details, width=20, height=2).pack(pady=5)
    Button(logg, text="Update Student Details", bg="white", fg="black", command=update_student_details, width=20, height=2).pack(pady=5)
    Button(logg, text="View Student Details", bg="white", fg="black", command=view_student_details, width=20, height=2).pack(pady=5)
    Button(logg, text="Delete Student Details", bg="white", fg="black", command=delete_student_details, width=20, height=2).pack(pady=5)


def failed():
    global fail
    fail = Toplevel(root2)
    fail.title("Invalid")
    fail.geometry("200x100")
    Label(fail, text="Invalid credentials...", fg="red", font="bold").pack()
    Label(fail, text="").pack()
    Button(fail, text="Ok", bg="white", fg="black", width=8, height=1, command=fail_destroy).pack()

def login_varify():
    user_varify = username_varify.get()
    pas_varify = password_varify.get()
    sql = "SELECT * FROM login WHERE username = %s AND password = %s"
    mycur.execute(sql, [(user_varify), (pas_varify)])
    results = mycur.fetchall()
    if results:
        for i in results:
            logged()
            break
    else:
        failed()

def add_student_details():
    global add_window
    add_window = Toplevel(logg)
    add_window.title("Please Add Your Student Details")
    add_window.geometry("300x300")

    global student_id_entry, first_name_entry, last_name_entry, address_entry, phone_number_entry, email_entry, dob_entry

    Label(add_window, text="Enter Student Details", font="bold").pack()
    Label(add_window, text="").pack()

    Label(add_window, text="Student ID: ").pack()
    student_id_entry = Entry(add_window)
    student_id_entry.pack()

    Label(add_window, text="First Name: ").pack()
    first_name_entry = Entry(add_window)
    first_name_entry.pack()

    Label(add_window, text="Last Name: ").pack()
    last_name_entry = Entry(add_window)
    last_name_entry.pack()

    Label(add_window, text="Address: ").pack()
    address_entry = Entry(add_window)
    address_entry.pack()

    Label(add_window, text="Phone Number: ").pack()
    phone_number_entry = Entry(add_window)
    phone_number_entry.pack()

    Label(add_window, text="Email: ").pack()
    email_entry = Entry(add_window)
    email_entry.pack()

    Label(add_window, text="Date of Birth: ").pack()
    dob_entry = Entry(add_window)
    dob_entry.pack()

    Label(add_window, text="").pack()

    Button(add_window, text="Submit", bg="white", fg="black", command=submit_student_details).pack()

def submit_student_details():
    student_id = student_id_entry.get()
    first_name = first_name_entry.get()
    last_name = last_name_entry.get()
    address = address_entry.get()
    phone_number = phone_number_entry.get()
    email = email_entry.get()
    dob = dob_entry.get()

    if student_id == "" or first_name == "" or last_name == "" or address == "" or phone_number == "" or email == "" or dob == "":
        messagebox.showerror("Error", "All fields are required")
    else:
        sql = "INSERT INTO student_details (student_id, first_name, last_name, address, phone_number, email, dob) VALUES (%s, %s, %s, %s, %s, %s, %s)"
        values = (student_id, first_name, last_name, address, phone_number, email, dob)
        mycur.execute(sql, values)
        db.commit()
        messagebox.showinfo("Success", "Student details added successfully!")
        add_window.destroy()

def view_student_details():
    def search_student():
        search_id = search_entry.get()
        if search_id == "":
            messagebox.showerror("Error", "Please enter a Student ID to search for details.")
        else:
            # Clear existing treeview entries
            for record in tree.get_children():
                tree.delete(record)

            # Fetch student details from the database for the provided ID
            mycur.execute("SELECT * FROM student_details WHERE student_id = %s", (search_id,))
            student_record = mycur.fetchone()
            if student_record:
                # Insert the fetched record into the treeview
                tree.insert("", "end", values=student_record)
            else:
                messagebox.showinfo("Info", "No student details found for the provided ID.")

    view_window = Toplevel(logg)
    view_window.title("View Student Details")
    view_window.geometry("500x300")

    # Search Frame
    search_frame = Frame(view_window)
    search_frame.pack(side="top", fill="x")

    Label(search_frame, text="Enter Student ID: ").pack(side="left")
    search_entry = Entry(search_frame)
    search_entry.pack(side="left", padx=5)
    search_button = Button(search_frame, text="Search", command=search_student)
    search_button.pack(side="left", padx=5)

    # Treeview
    tree = ttk.Treeview(view_window, columns=(1, 2, 3, 4, 5, 6, 7), show="headings", height="10")
    tree.pack(side='left')

    tree.heading(1, text="Student ID")
    tree.heading(2, text="First Name")
    tree.heading(3, text="Last Name")
    tree.heading(4, text="Address")
    tree.heading(5, text="Phone Number")
    tree.heading(6, text="Email")
    tree.heading(7, text="Date of Birth")

    # Add scrollbar
    scrollbar = ttk.Scrollbar(view_window, orient="vertical", command=tree.yview)
    scrollbar.pack(side='right', fill='y')
    tree.configure(yscrollcommand=scrollbar.set)


def update_student_details():
    global update_window
    update_window = Toplevel(logg)
    update_window.title("Update Student Details")
    update_window.geometry("300x300")

    global update_student_id_entry, update_first_name_entry, update_last_name_entry, update_address_entry, update_phone_number_entry, update_email_entry, update_dob_entry

    Label(update_window, text="Enter Student ID to Update Details", font="bold").pack()
    Label(update_window, text="").pack()

    Label(update_window, text="Student ID: ").pack()
    update_student_id_entry = Entry(update_window)
    update_student_id_entry.pack()

    Button(update_window, text="Search", bg="blue", command=search_student_details).pack()

def search_student_details():
    student_id = update_student_id_entry.get()
    if student_id == "":
        messagebox.showerror("Error", "Please enter Student ID to search for details.")
    else:
        sql = "SELECT * FROM student_details WHERE student_id = %s"
        mycur.execute(sql, (student_id,))
        record = mycur.fetchone()
        if record:
            show_update_window(record)
        else:
            messagebox.showinfo("Info", "No student details found for the provided ID.")

def show_update_window(record):
    global update_window
    update_window = Toplevel(logg)
    update_window.title("Update Student Details")
    update_window.geometry("300x300")

    global update_first_name_entry, update_last_name_entry, update_address_entry, update_phone_number_entry, update_email_entry, update_dob_entry

    Label(update_window, text="Update Student Details", font="bold").pack()
    Label(update_window, text="").pack()

    Label(update_window, text="First Name: ").pack()
    update_first_name_entry = Entry(update_window)
    update_first_name_entry.insert(0, record[1])
    update_first_name_entry.pack()

    Label(update_window, text="Last Name: ").pack()
    update_last_name_entry = Entry(update_window)
    update_last_name_entry.insert(0, record[2])
    update_last_name_entry.pack()

    Label(update_window, text="Address: ").pack()
    update_address_entry = Entry(update_window)
    update_address_entry.insert(0, record[3])
    update_address_entry.pack()

    Label(update_window, text="Phone Number: ").pack()
    update_phone_number_entry = Entry(update_window)
    update_phone_number_entry.insert(0, record[4])
    update_phone_number_entry.pack()

    Label(update_window, text="Email: ").pack()
    update_email_entry = Entry(update_window)
    update_email_entry.insert(0, record[5])
    update_email_entry.pack()

    Label(update_window, text="Date of Birth: ").pack()
    update_dob_entry = Entry(update_window)
    update_dob_entry.insert(0, record[6])
    update_dob_entry.pack()

    Label(update_window, text="").pack()

    Button(update_window, text="Update", bg="white", fg="black", command=submit_update_student_details).pack()

def submit_update_student_details():
    student_id = update_student_id_entry.get()
    first_name = update_first_name_entry.get()
    last_name = update_last_name_entry.get()
    address = update_address_entry.get()
    phone_number = update_phone_number_entry.get()
    email = update_email_entry.get()
    dob = update_dob_entry.get()

    if first_name == "" or last_name == "" or address == "" or phone_number == "" or email == "" or dob == "":
        messagebox.showerror("Error", "All fields are required")
    else:
        sql = "UPDATE student_details SET first_name = %s, last_name = %s, address = %s, phone_number = %s, email = %s, dob = %s WHERE student_id = %s"
        values = (first_name, last_name, address, phone_number, email, dob, student_id)
        mycur.execute(sql, values)
        db.commit()
        messagebox.showinfo("Success", "Student details updated successfully!")
        update_window.destroy()

def delete_student_details():
    delete_window = Toplevel(logg)
    delete_window.title("Please Delete Your Student Details")
    delete_window.geometry("300x100")

    def delete_records():
        sql = "DELETE FROM student_details"
        mycur.execute(sql)
        db.commit()
        messagebox.showinfo("Success", "All student details deleted successfully!")
        delete_window.destroy()

    Label(delete_window, text="Are you sure you want to delete all student details?", font="bold").pack()
    Label(delete_window, text="").pack()
    Button(delete_window, text="Yes", bg="red", fg="black", command=delete_records).pack(side=LEFT, padx=10)
    Button(delete_window, text="No", bg="white", fg="black", command=delete_window.destroy).pack(side=RIGHT, padx=10)


def main_screen():
    global root
    root = Tk()
    root.title("University Registration System")
    root.geometry("300x300")

    Label(root, text="University Registration System", font=("Arial", 14, "bold")).pack()
    Label(root, text="").pack()

    Label(root, text="").pack()
    Button(root, text="Log-IN", width="15", height="2", bg="white", fg="black", command=login).pack()
    Label(root, text="").pack()
    Button(root, text="Registration", height="2", width="15", bg="white", fg="black", command=registration).pack()
    Label(root, text="").pack()
    Label(root, text="").pack()
    Label(root, text="Developed By Ashok Kasaram").pack()

main_screen()
root.mainloop()
```
